```python
from statsforecast import StatsForecast
from statsforecast.models import (
    AutoARIMA, ETS, HistoricAverage,
    Naive, CrostonClassic, IMAPA, SeasonalNaive
)
```

```python
import pickle
import pandas as pd
import os

download_path = os.path.abspath(os.path.join(".","..","data","m5-forecasting-accuracy.zip"))
unzipped_path = os.path.abspath(os.path.join(".","..","data","m5-forecasting-accuracy-unzipped"))

# Read in the data
INPUT_DIR = unzipped_path
WORKING_DIR = os.path.join(unzipped_path, "..", "working")
```

```python
df = pd.read_parquet(f"{WORKING_DIR}/series.parquet")
df = df.iloc[0]
new = pickle.loads(df['series'])
new['unique_id'] = df['id']
new = new.reset_index()
new.columns = ["ds", "y", "price", "unique_id"]
new['y'] = new['y'] + 0.01
```

```python
fcst = StatsForecast(
    df=new,
    models=[Naive()],
    freq='D',
)
```

```python
fcst.forecast(h=28)
```

```python
models_to_fit = [
    Naive(), 
    AutoARIMA(),
    ETS(season_length=7), 
    CrostonClassic(),
    IMAPA()
]

fcst = StatsForecast(
    df=new,
    models=models_to_fit,
    freq='D',
)
```

```python
fcst.forecast(28)
```
